# 任务描述
通过高德地图 Web API 爬取广州市 8 个行政区、5 个主流便利店品牌的真实门店数据，获取门店名称、地址、品牌、区域、经纬度信息，为选址分析提供原始数据支撑。

```python
# 01_crawl_convenience_stores.py
# 完整版：一次爬完广州8区+5大品牌 | 全程打印日志 | 绝不卡死
import pandas as pd
import requests
import time

# ====================== 填入你的高德Web服务KEY ======================
GAODE_API_KEY = "c9388e2654c8ecac4ed8e8080916364c"

# ====================== 全广州 8 区 + 5 大品牌 ======================
BRANDS = ["美宜佳", "全家", "7-Eleven", "喜士多", "天福"]
DISTRICTS = ["天河区", "越秀区", "荔湾区", "海珠区", "白云区", "黄埔区", "番禺区", "花都区"]

all_stores = []

# ====================== 开始爬取（全程打印，绝不静默） ======================
print("=" * 60)
print("  广州便利店全量爬取开始")
print(f"  区域：{len(DISTRICTS)} 个区")
print(f"  品牌：{len(BRANDS)} 个")
print("=" * 60)

for brand in BRANDS:
    for district in DISTRICTS:
        page = 1
        print(f"\n📍 正在爬取：【{district}】|【{brand}】")

        while True:
            url = (
                f"https://restapi.amap.com/v3/place/text"
                f"?keywords={brand}"
                f"&city=广州"
                f"&region={district}"
                f"&offset=20"
                f"&page={page}"
                f"&key={GAODE_API_KEY}"
                f"&output=json"
            )

            # 超时 5 秒，不会卡死
            try:
                res = requests.get(url, timeout=5)
                data = res.json()
            except Exception as e:
                print(f"   ⚠️  第 {page} 页请求失败，跳过")
                break

            # API 错误
            if data.get("status") != "1":
                print(f"   ⚠️  API 错误：{data.get('info')}")
                break

            # 没有数据了
            if len(data.get("pois", [])) == 0:
                print(f"   ✅ {district} {brand} 无更多数据")
                break

            # 读取数据
            count = 0
            for p in data["pois"]:
                try:
                    lon, lat = p["location"].split(",")
                    all_stores.append({
                        "name": p["name"],
                        "brand": brand,
                        "district": p["adname"],
                        "address": p["address"],
                        "longitude": float(lon),
                        "latitude": float(lat)
                    })
                    count += 1
                except:
                    continue

            print(f"   📄 第 {page} 页 → 获取 {count} 条")
            page += 1
            time.sleep(0.5)

# ====================== 保存最终数据 ======================
df = pd.DataFrame(all_stores)
df = df.drop_duplicates(subset=["name", "address"])
df.to_csv("data/raw_stores.csv", index=False, encoding="utf-8-sig")

print("\n" + "=" * 60)
print("🎉 全部爬取完成！")
print(f"✅ 总有效店铺：{len(df)} 家")
print("✅ 文件已保存到：data/raw_stores.csv")
print("=" * 60)
# 保存已爬到的数据并结束
df = pd.DataFrame(all_stores)
df = df.drop_duplicates(subset=["name", "address"])
df.to_csv("data/raw_stores.csv", index=False, encoding="utf-8-sig")

print("\n" + "=" * 60)
print("✅ 手动保存成功！")
print(f"✅ 当前已爬取店铺数量：{len(df)} 家")
print("✅ 文件路径：data/raw_stores.csv")
print("=" * 60)

# 退出，不继续爬
import sys
sys.exit()
```

```python
# 01_crawl_convenience_stores.py（最终无错版）
import pandas as pd
import requests
import time

GAODE_API_KEY = "c9388e2654c8ecac4ed8e8080916364c"
BRANDS = ["美宜佳", "全家", "7-Eleven", "喜士多", "天福"]
DISTRICTS = ["天河区", "越秀区", "荔湾区", "海珠区", "白云区", "黄埔区", "番禺区", "花都区"]

all_stores = []
print("=" * 60)
print("  广州便利店全量爬取开始")
print(f"  区域：{len(DISTRICTS)} 个区")
print(f"  品牌：{len(BRANDS)} 个")
print("=" * 60)

for brand in BRANDS:
    for district in DISTRICTS:
        page = 1
        print(f"\n📍 正在爬取：【{district}】|【{brand}】")

        while True:
            url = (
                f"https://restapi.amap.com/v3/place/text"
                f"?keywords={brand}"
                f"&city=广州"
                f"&region={district}"
                f"&offset=20"
                f"&page={page}"
                f"&key={GAODE_API_KEY}"
                f"&output=json"
            )

            try:
                res = requests.get(url, timeout=5)
                data = res.json()
            except Exception as e:
                print(f"   ⚠️  第 {page} 页请求失败，跳过")
                break

            if data.get("status") != "1":
                print(f"   ⚠️  API 错误：{data.get('info')}")
                break

            if len(data.get("pois", [])) == 0:
                print(f"   ✅ {district} {brand} 无更多数据")
                break

            count = 0
            for p in data["pois"]:
                try:
                    # 强制转字符串，彻底杜绝list类型
                    name = str(p.get("name", ""))
                    district_name = str(p.get("adname", ""))
                    address = str(p.get("address", ""))
                    location = p.get("location", "0,0")
                    
                    # 兼容经纬度的两种格式
                    if isinstance(location, list):
                        lon, lat = str(location[0]), str(location[1])
                    else:
                        lon, lat = location.split(",")
                    
                    all_stores.append({
                        "name": name,
                        "brand": brand,
                        "district": district_name,
                        "address": address,
                        "longitude": float(lon),
                        "latitude": float(lat)
                    })
                    count += 1
                except:
                    continue

            print(f"   📄 第 {page} 页 → 获取 {count} 条")
            page += 1
            time.sleep(0.5)

# 最终安全保存
df = pd.DataFrame(all_stores)
df = df.astype(str).drop_duplicates(subset=["name", "address"], keep="first")
df["longitude"] = df["longitude"].astype(float)
df["latitude"] = df["latitude"].astype(float)
df.to_csv("data/raw_stores.csv", index=False, encoding="utf-8-sig")

print("\n" + "=" * 60)
print("🎉 全部爬取完成！")
print(f"✅ 总有效店铺：{len(df)} 家")
print("✅ 文件已保存到：data/raw_stores.csv")
print("=" * 60)
```
