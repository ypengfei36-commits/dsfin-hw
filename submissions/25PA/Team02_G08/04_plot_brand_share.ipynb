## 4. 品牌竞争格局可视化（04_plot_brand_share.ipynb）
### 任务目标
基于数据清洗后生成的品牌统计数据，绘制广州市便利店品牌门店数量分布柱状图，直观呈现各品牌的市场规模差异，分析品牌竞争格局。

### 输入数据
- `data/brand_stats.csv`：包含各品牌门店数量的统计结果

### 核心处理步骤
1.  读取品牌统计数据，按门店数量从高到低排序
2.  绘制柱状图，标注各品牌的真实门店数量
3.  优化图表标注、配色与排版，符合学术报告规范
4.  输出高分辨率PNG图表，供汇报使用

### 输出文件
- `output/fig2_brand.png`：广州市便利店品牌门店数量分布图

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

# 创建输出文件夹
os.makedirs("output", exist_ok=True)

# 读取真实品牌统计数据，自动过滤空值
df = pd.read_csv("data/brand_stats.csv")
# 核心修复：删除包含NaN的无效行
df = df.dropna(subset=["brand", "shop_count"]).copy()
# 过滤小众品牌（仅保留门店数≥10的品牌，保证图表可读性）
df = df[df["shop_count"] >= 10].sort_values("shop_count", ascending=False).reset_index(drop=True)

# 学术风格绘图配置
plt.rcParams["font.sans-serif"] = ["SimHei", "Microsoft YaHei", "Times New Roman"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.facecolor"] = "white"

# 绘制品牌门店数量柱状图
fig, ax = plt.subplots(figsize=(12, 7))
bars = ax.bar(
    df["brand"], 
    df["shop_count"], 
    color="#F18F01",  # 学术暖橙，避免花哨
    alpha=0.8,
    width=0.6
)

# 安全处理数值标签
for bar in bars:
    height = bar.get_height()
    if not np.isnan(height):
        ax.text(
            bar.get_x() + bar.get_width()/2,
            height + 3,  # 标签偏移，避免遮挡
            f"{int(round(height))}",
            ha="center", 
            va="bottom", 
            fontsize=11,
            fontweight="bold"
        )

# 图表标注
ax.set_title("广州市便利店品牌门店数量分布", fontsize=16, fontweight="bold", pad=20)
ax.set_xlabel("便利店品牌", fontsize=13, fontweight="bold")
ax.set_ylabel("门店数量（家）", fontsize=13, fontweight="bold")
ax.set_ylim(0, max(df["shop_count"]) * 1.15)
ax.grid(axis="y", alpha=0.3, linestyle="-")

# 调整布局，避免标签截断
plt.tight_layout()

# 保存高分辨率图片
plt.savefig("output/fig2_brand.png", dpi=300, bbox_inches="tight", facecolor="white")
plt.close()

# 输出运行日志
print("✅ 品牌格局图生成完成：output/fig2_brand.png")
print("\n=== 品牌门店数量统计（Top10）===")
print(df.head(10)[["brand", "shop_count"]])
```
